```python
import numpy as np
import matplotlib.pyplot as plt
a0 = 1.0
omega0 = 1.02
H0 = 70
x = np.linspace(0,6.2)
a = a0*omega0*(1-np.cos(x))/(2*(omega0-1))
plt.plot(x,a,label=f"$a_0=1.0, \Omega_0=1.02$")
plt.xlabel(f"$\Theta$")
plt.ylabel(f"$a(\Theta)$")
plt.legend(loc=(0,1));
```

```python
import numpy as np
import matplotlib.pyplot as plt
a0 = 1.0
omega0 = 1.02
H0 = 70
x = np.linspace(0,6.2)
t = omega0*(x-np.sin(x))/(2*H0*(omega0-1)**(3/2))
plt.plot(x,t,label=f"$a_0=1.0, \Omega_0=1.02, H0=70km/s/Mpc$")
plt.xlabel(f"$\Theta$")
plt.ylabel(f"$t(\Theta)(Gyrs)$")
plt.legend(loc=(0,1));
```
