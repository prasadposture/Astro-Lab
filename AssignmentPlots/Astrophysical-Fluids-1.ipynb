```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# Define velocity components with t = 2
fig, axes = plt.subplots(2,3, figsize=(20, 10))
t=1
for i in [0, 1]:
    for j in [0,1,2]:
        U = 2*X*Y
        V = t*Y**2
        axes[i,j].streamplot(X, Y, U, V, color='b', linewidth=2, density=1)
        axes[i,j].scatter(x=2,y=-1, color='r', s=100)
        # Set plot title and labels
        axes[i,j].set_title('Streamlines with Velocity Vector at t='+str(t))
        axes[i,j].set_xlabel('X')
        axes[i,j].set_ylabel('Y')
        t=t*2
plt.savefig('P3.1.1.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# Define velocity components with t = 2
fig, axes = plt.subplots(2,3, figsize=(20, 10))
t=1
for i in [0, 1]:
    for j in [0,1,2]:
        U = 2 * Y*(X-Y)
        V = X*Y*t
        axes[i,j].streamplot(X, Y, U, V, color='b', linewidth=2, density=1)
        axes[i,j].scatter(x=2,y=-1, color='r', s=100)
        # Set plot title and labels
        axes[i,j].set_title('Streamlines with Velocity Vector at t='+str(t))
        axes[i,j].set_xlabel('X')
        axes[i,j].set_ylabel('Y')
        t=t*2
plt.savefig('P3.2.1.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta_values = np.linspace(0, 2 * np.pi, 100)
c_prime = 1  # Modify this value as needed

# Calculate r values based on the equation
r_values = c_prime * np.sin(theta_values / 2)

plt.figure(figsize=(8, 8))
plt.polar(theta_values, r_values, label=r'$r = c\' \sin(\frac{\theta}{2})$')
plt.title(r'Polar Plot: $r = c\' \sin(\frac{\theta}{2})$')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('P6.3.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the radius of the circle
R = 1.5
k=1
c=1
# Generate values for theta

#theta = np.linspace(0, 2*np.pi, 100)
for nt in [2,4,6,8,10,12]:
    t = np.linspace(0, nt, 100)
    # Calculate radial distance
    r = np.ones_like(t) * R
    theta = (k * t) / (r ** 2) + c
    # Plot the polar plot
    plt.figure(figsize=(6, 6))
    plt.polar(theta, r)
    plt.title(f'Polar Plot of a Circle with  at t={nt}s')
    plt.show()
```
